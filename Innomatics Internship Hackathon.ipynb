### Mandatory operations

```python
movies_ratings
```

```python
#To group the "movieId" and "ratings" (SUM)
suma = movies_ratings.groupby("movieId")["rating"].sum()
sum1 = pd.DataFrame(suma)
sum1
```

```python
#Count of "moviesId" by "ratings" (COUNT)
counta = movies_ratings.groupby("movieId")["rating"].value_counts()
count1 = pd.DataFrame(counta)
count1
```

```python
#Count of "moviesId" by "ratings" (MEAN)
mean_a = movies_ratings.groupby("movieId")["rating"].mean()
mean1 = pd.DataFrame(mean1)
mean1
```

### Now perform an inner join

```python
movies
```

```python
#For the 1st for count
ratings_count = movies.merge(count1, how = "inner", on = "movieId")
ratings_count
```

```python
#For the 2nd for mean
ratings_mean = mean1.merge(movies, on = "movieId", how = "inner")
ratings_mean
```

### Now filter only movies with ratings greater than 50

```python
filter_count = ratings_count.loc[ratings_count["rating"] > 50]
filter_count
```

```python
#This is just to filter the movies with top average user ratings too
filter_mean = ratings_mean.loc[ratings_mean["rating"] > 3]
filter_mean
```

```python
#For the movie with the highest average user rating, do this
filter_count.groupby("title")["rating"].sum().sort_values(ascending = False)
```
